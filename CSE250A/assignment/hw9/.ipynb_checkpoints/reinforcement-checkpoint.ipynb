```python
import os
import numpy as np
import pandas as pd
```

```python
#load data files
def readfile(file_path):
    data = np.zeros((81,81))
    with open(file_path) as f:
        content = f.readlines()
    for i in range(len(content)):
        r,c,value = [x for x in content[i].strip().split(" ") if x is not ""]
        data[int(r)-1,int(c)-1] = float(value)
    return data

def readReward(file_path):
    reward = np.zeros(81)
    with open(file_path) as f:
        content = f.readlines()
    for i in range(len(content)):
        reward[i] = int(content[i].strip())
    return reward
a1 = readfile("prob_a1.txt")
a2 = readfile("prob_a2.txt")
a3 = readfile("prob_a3.txt")
a4 = readfile("prob_a4.txt")
reward = readReward("rewards.txt")
```

```python
def max_value_action(action, state, v_pre):
    max_value  = -np.inf
    max_action = -np.inf
    for a_index in range(4):
        temp = 0.0
        for s_next in range(81):
            temp += action[a_index][state,s_next] * v_pre[s_next]
        if temp > max_value:
            max_value = temp
            max_action = a_index
    return max_value, max_action


def get_matrix_policy(action, pi):
    T = np.zeros((81, 81))
    for s in range(81):
        next_a = int(pi[s])
        T[s,:] = action[next_a][s,:]
    return T

def policy_iteration(a, rewards, V):
    pi = np.zeros(81)
    gamma = 0.99
    I = np.eye(81)
    for _ in range(100):
        P = get_matrix_policy(a, pi)
        v_pre = np.linalg.inv(I - gamma * P).dot(rewards)
        for s in range(81):
            max_value, max_action = max_value_action(a, s, v_pre)
            V[s] = rewards[s] + gamma * max_value
            pi[s] = max_action
    return V, pi

def value_iteration(a, rewards, V):
    pi = np.zeros(81)
    gamma = 0.99
    I = np.eye(81)
    for _ in range(1000):
        for s in range(81):
            max_value, max_action = max_value_action(a, s, V)
            V[s] = rewards[s] + gamma * max_value
            pi[s] = max_action
    return V, pi
```

# Policy Iteration

```python
V = np.zeros(81)
a = [a1, a2, a3, a4]
V, PI = policy_iteration(a, reward, V)

maze = np.zeros((9,9))
for i in range(9):
    for j in range(9):
        maze[i,j] = round(V[i+j*9],2)
value_show = pd.DataFrame({"0":maze[:,0],"1":maze[:,1],"2":maze[:,2],
                         "3":maze[:,3],"4":maze[:,4],"5":maze[:,5],
                         "6":maze[:,6],"7":maze[:,7],"8":maze[:,8]})
value_show
```

```python
PI_M = PI.reshape((9,9)).T.astype(np.int)
PI_show = pd.DataFrame({"0":PI_M[:,0],"1":PI_M[:,1],"2":PI_M[:,2],
                         "3":PI_M[:,3],"4":PI_M[:,4],"5":PI_M[:,5],
                         "6":PI_M[:,6],"7":PI_M[:,7],"8":PI_M[:,8]})
PI_show
```

# Value Iteration

```python
V_2 = np.zeros(81)
a = [a1, a2, a3, a4]
V_2, PI_2 = value_iteration(a, reward, V_2)

maze_2 = np.zeros((9,9))
for i in range(9):
    for j in range(9):
        maze_2[i,j] = round(V_2[i+j*9],2)
value_show_2 = pd.DataFrame({"0":maze_2[:,0],"1":maze_2[:,1],"2":maze_2[:,2],
                         "3":maze_2[:,3],"4":maze_2[:,4],"5":maze_2[:,5],
                         "6":maze_2[:,6],"7":maze_2[:,7],"8":maze_2[:,8]})
value_show_2
```
